## Changing Native Country Column to BP, LP, US

```python
def native_country(text):
    if text in bp:
        return 'BP'
    elif text in base:
        return 'UnitedStates'
    elif text in lp:
        return 'LP'
```

```python
train['country'] = train['native-country'].apply(native_country)
```

```python
train.head()
```

```python
train['country'].value_counts()
```

## Changing Capital Gain and Capital Loss into binary

```python
def cap_gain(number):
    if number == 0:
        return 0
    else:
        return 1
    
def cap_loss(number): 
    if number == 0:
        return 0
    else: 
        return 1
```

```python
train['cap_gain'] = train['capital-gain'].apply(cap_gain)
train['cap_loss'] = train['capital-loss'].apply(cap_loss)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import BernoulliNB
```

## Logistic GridSearch

```python
pipe = Pipeline([
    ('kbest', SelectKBest()),
    ('lr', LogisticRegression())
])
pipe_params = {
    'kbest__k': [15, 16, 17, 18, 19, 20],
    'lr__C': [0.2, 0.4, 0.6, 0.8, 1 ],
    'lr__penalty': ['l1', 'l2']
    #'rf__max_features': ['auto', 'sqrt']
    #can add C parameter for logisitic regression
}


gs = GridSearchCV(pipe, param_grid=pipe_params, cv=3)


gs.fit(X_train, y_train)


print(gs.best_score_)
gs.best_params_
```

```python
gs.score(X_train, y_train)
```

```python
gs.score(X_test, y_test)
```

```python
pred = gs.predict_proba(X_test)
```

```python
log_loss(y_test, pred)
```
